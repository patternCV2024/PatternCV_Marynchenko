# Lucky series

```python
# lucky series

import numpy as np

def lucky(num: str):
    lucky_numbers = np.array(['5','6'])
    num = str(num)
    i = ''
    series_len = []
    for s in num:
        if s in lucky_numbers:
            i += s
        else:
            if np.array_equal(
                np.intersect1d(
                    np.array(list(i)), lucky_numbers
                ), lucky_numbers
            ):
                series_len.append(len(i))
            i = ''

    return max(series_len) if series_len else len(
        i if np.array_equal(
                np.intersect1d(
                    np.array(list(i)), lucky_numbers
                ), lucky_numbers
            ) else ''
    )
```

```python
lucky('555555556')
```

```python
lucky('55555555')
```

```python
lucky('4556432455665334')
```
